```python
# bring libraries into scope
import numpy as np
import math
import sys
import pandas as pd

# set print options
np.set_printoptions(threshold=np.nan)

# import data
p2file = "/Users/C3P0/Code/OMSCS/CS7646_ML4T/Projects/MC3_P1/mc3p1_assess_learners/Istanbul.csv"
inf = open(p2file)
data = np.array([map(float,s.strip().split(',')[1:]) for s in inf.readlines()[1:]])

# compute how much of the data is training and testing
train_rows = int(0.6* data.shape[0])
test_rows = data.shape[0] - train_rows

# separate out training and testing data
trainX = data[:train_rows,0:-1]
trainY = data[:train_rows,-1]
testX = data[train_rows:,0:-1]
testY = data[train_rows:,-1]

# rename train data as dataX and DataY to develop final function
dataX = trainX
dataY = trainY

# set seed
np.random.seed(2)
```

```python
def buildTree(dataX, dataY, leafsize):
    if dataX.shape[0] <= leafsize:
        return np.array([[-1, dataY.mean(), np.nan, np.nan]])
    elif dataY.std() == 0:
        return np.array([[-1, dataY.mean(), np.nan, np.nan]])
    else:
        i, splitVal, leftX, rightX, leftY, rightY, leftleaf = findNonEmptyRandomSplit(dataX, dataY)
        
        if leftleaf.size == 0:
            lefttree = buildTree(leftX, leftY, leafsize)
            righttree = buildTree(rightX, rightY, leafsize)
            root = np.array([[i, splitVal, 1, lefttree.shape[0] + 1]])
            fin_array = np.append(root, lefttree, axis=0)
            fin_array = np.append(fin_array, righttree, axis=0)
            return fin_array
        else:
            lefttree = np.append(lefttree, leftleaf, axis=0)
            righttree = buildTree(rightX, rightY, leafsize)
            root = np.array([[i, splitVal, 1, lefttree.shape[0] + 1]])
            fin_array = np.append(root, lefttree, axis=0)
            fin_array = np.append(fin_array, righttree, axis=0)
            return fin_array            
```

```python
def findNonEmptyRandomSplit(dataX, dataY):
    counter = 0   # start counter
    
    while counter <= 10:
        counter = counter + 1   # increment counter
        
        i = np.random.randint(0, dataX.shape[1])    # randomly select a feature
        j = np.random.randint(0, dataX.shape[0])    # randomly select element 1 in i
        k = np.random.randint(0, dataX.shape[0])    # randomly select element 2 in i
        
        splitVal = (dataX[j, i] + dataX[k, i]) / 2  # find the value to split on
        
        # build left and right X and Y based off of splitVal
        leftIndex = np.where(dataX[:,i] <= splitVal)
        rightIndex = np.where(dataX[:,i] > splitVal)
        leftX = dataX[leftIndex]
        leftY = dataY[leftIndex]
        rightX = dataX[rightIndex]
        rightY = dataY[rightIndex]
        
        if leftX.size == 0:
            if counter == 10:
                leaf = np.array([[-1, leftY.mean(), np.nan, np.nana]])
                return i, splitVal, leftX, rightX, leftY, rightY, leaf
            else:
                pass
        else:
            leaf = np.array([])
            return i, splitVal, leftX, rightX, leftY, rightY, leaf
```

```python
def queryPT(pt, model):
    index = 0
    node = model[index]
    feature, splitVal, indexL, indexR = node
    feature = int(feature)
    indexL = int(indexL)
    indexR = int(indexR)
    while feature != -1:
        if pt[feature] <= splitVal:
            index = index + indexL
            node = model[index]
            feature, splitVal, indexL, indexR = node
            feature = int(feature)
            try:
                indexL = int(indexL)
                indexR = int(indexR)
            except ValueError:
                pass
            if feature == -1:
                return splitVal
            else:
                pass
        else:
            index = index + indexR
            node = model[index]
            feature, splitVal, indexL, indexR = node
            feature = int(feature)
            try:
                indexL = int(indexL)
                indexR = int(indexR)
            except ValueError:
                pass
            if feature == -1:
                return splitVal
            else:
                pass
```
